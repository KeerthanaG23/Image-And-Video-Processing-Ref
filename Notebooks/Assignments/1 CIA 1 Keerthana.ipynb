#### Approximate distribution means that we assume certain discrete values for the heights and weights of the students & plot the distribution based on the assumed values i.e the discrete points provided

#### Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Creating bins to categorise all the continous heights and weights

```python
bin_edges_1 = np.array([5.4,5.6,5.8,6])
bin_edges_2 = np.array([50,60,65,70,75])
```

```python
print(type(bin_edges_1))
print(type(bin_edges_2))
```

#### Generating random weights and heights within the given range for 70 students

```python
# Defining range
a,b = (5.4,6)
c,d = (50,75)
```

```python
def h_w(a,b):
    param = []
    
    for i in range(70):
        param.append(round(np.random.uniform(a,b),2))     
    return param
```

```python
heights=np.array(h_w(a,b))
```

```python
weights=np.array(h_w(c,d))
```

```python
heights
```

#### Based on the random heights and weights generating estimating the count of students within each range

```python
histogram_counts_1,_= np.histogram(heights,bins=bin_edges_1)
histogram_counts_2,_= np.histogram(weights,bins=bin_edges_2)
```

```python
histogram_counts_1=histogram_counts_1.tolist()
print(histogram_counts_1)
```

```python
histogram_counts_2=histogram_counts_2.tolist()
print(histogram_counts_2)
```

### Plotting the approximate height distribution 

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Height Distribution')
plt.xlabel('Height (ft)')
plt.ylabel('Frequency')
plt.hist(heights,bins=bin_edges_1,edgecolor="black")
```

### Plotting the approximate weight distribution

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.hist(weights,bins=bin_edges_2,edgecolor="black",color="red")
```

```python
plt.figure(figsize=(9, 4))
plt.subplot(1, 2, 1)
plt.scatter(heights, weights, alpha=0.7, s=30 * np.ones_like(heights), c=heights, cmap='viridis')
plt.title('Joint Distribution of Height and Weight')
plt.xlabel('Height (ft)')
plt.ylabel('Weight (kg)')
cbar = plt.colorbar()
cbar.set_label('Height (ft)')
plt.show()
```

#### Calculate joint probabilities using a bivariate normal distribution

```python
from scipy.stats import multivariate_normal
```

```python

mean = [np.mean(heights), np.mean(weights)]
covariance_matrix = np.cov(heights, weights)
joint_probabilities = multivariate_normal.pdf(np.column_stack((heights, weights)), mean, covariance_matrix)
```

```python
plt.figure(figsize=(9, 4))


plt.subplot(1, 2, 1)
plt.scatter(heights, weights, alpha=0.7, s=1000 * joint_probabilities, c=heights, cmap='viridis')
plt.title('Joint Distribution of Height and Weight')
plt.xlabel('Height (ft)')
plt.ylabel('Weight (kg)')
cbar = plt.colorbar()
cbar.set_label('Height (ft)')


plt.subplot(1, 2, 2)
plt.scatter(heights, weights, alpha=0.7, s=1000 * joint_probabilities, c=weights, cmap='viridis')
plt.title('Joint Distribution of Height and Weight')
plt.xlabel('Height (ft)')
plt.ylabel('Weight (kg)')
cbar = plt.colorbar()
cbar.set_label('Weight (kg)')

plt.tight_layout()
plt.show()
```

### Skewness & Kurtosis

```python
from scipy.stats import skew, kurtosis, mode,norm
```

```python
skewness_value = skew(heights)
kurtosis_value = kurtosis(heights)

plt.hist(heights, bins=20, edgecolor='black', alpha=0.7, label='Height Distribution')

plt.axvline(mean_h, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_h, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(mode_h, color='blue', linestyle='dashed', linewidth=2, label='Mode')

plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height Distribution with Mean, Median, and Mode')
plt.legend()
plt.show()

# Print skewness and kurtosis
print("Skewness:", skewness_value)
print("Kurtosis:", kurtosis_value)




# Fit a Gaussian distribution
mu, std = norm.fit(heights)

# Plot the histogram
plt.hist(heights, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black', label='Height Distribution')

# Plot the Gaussian distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Gaussian Fit')

# Add labels and a title
plt.xlabel('Height')
plt.ylabel('Density')
plt.title('Height Distribution with Gaussian Fit')
plt.legend()
plt.show()

# Print skewness and kurtosis
print("Skewness:", skewness_value)
print("Kurtosis:", kurtosis_value)
```

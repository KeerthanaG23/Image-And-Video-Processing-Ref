### Histogram of Negative image

```python
import matplotlib.pyplot as plt 
  
height = [5,2,3,4,0,7,0,5,6,6,4,7,7,6,6,5,4,5,3,1,
6,5,5,3,7] 
plt.title("Histogram of negative Image")
plt.xlabel("Grey Value")
plt.ylabel("Frequency")
plt.hist(height, edgecolor="white", bins=8) 
plt.show() 
```

### Original Image histogram 

```python
import matplotlib.pyplot as plt 
  
height = [2,5,4,3,7,0,7,2,1,1,3,0,0,1,1,2,3,2,4,6,1,2,2,4,0] 
plt.title("Histogram of original Image")
plt.xlabel("Grey Value")
plt.ylabel("Frequency")
plt.hist(height, edgecolor="white", bins=8) 
plt.show() 
```

### Equalized Image Histogram

```python
import matplotlib.pyplot as plt 
  
height = [4,6,6,5,7,1,7,4,3,3,5,1,1,3,3,4,5,4,6,6,3,4,4,6,1] 
plt.title("Histogram of Equalized Image")
plt.xlabel("Grey Value")
plt.ylabel("Frequency")
plt.hist(height, edgecolor="white", bins=[0,1,2,3,4,5,6,7,8]) 
plt.show() 
```

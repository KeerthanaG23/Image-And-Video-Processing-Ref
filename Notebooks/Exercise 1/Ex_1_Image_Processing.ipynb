```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the original image
img = cv2.imread(r"C:\Users\Keerthana\Desktop\image_new.jpeg")

# Convert the image to HSV color space
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds for blue color
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

# Define the lower and upper bounds for red color
lower_red1 = np.array([0, 50, 50])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])

# Create masks for blue and red regions
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_red1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask_red2 = cv2.inRange(hsv, lower_red2, upper_red2)
mask_red = mask_red1 + mask_red2  # Combine masks for red color

# Combine the blue and red masks
mask_boundary_boxes = cv2.bitwise_or(mask_blue, mask_red)

# Apply the mask to the original image
result = cv2.bitwise_and(img, img, mask=mask_boundary_boxes)

# Display the original image and the masked result using Matplotlib
plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(mask_boundary_boxes, cmap='gray')
plt.title('Blue and Red Masks'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Result with Masks'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the original image
img = cv2.imread(r"C:\Users\Keerthana\Desktop\image_new.jpeg")

# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img, (5, 5), 0)

# Extract individual color channels
b, g, r = cv2.split(img_blur)

# Apply Canny Edge Detection to each color channel
edges_b = cv2.Canny(b, 50, 150)
edges_g = cv2.Canny(g, 50, 150)
edges_r = cv2.Canny(r, 50, 150)

# Combine the Canny edges from each channel
edges_combined = edges_b | edges_g | edges_r

# Display the original image, Canny edges, and combined edges using Matplotlib
plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(edges_combined, cmap='gray')
plt.title('Combined Canny Edges'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the original image
img = cv2.imread(r"C:\Users\Keerthana\Desktop\image_new.jpeg")

# Blur the image
img_blur = cv2.GaussianBlur(img, (15, 15), 0)

# Convert the image to HSV color space
hsv = cv2.cvtColor(img_blur, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds for blue color
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

# Define the lower and upper bounds for red color
lower_red1 = np.array([0, 50, 50])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])

# Create masks for blue and red regions
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_red1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask_red2 = cv2.inRange(hsv, lower_red2, upper_red2)
mask_red = mask_red1 + mask_red2  # Combine masks for red color

# Combine the blue and red masks
mask_combined = cv2.bitwise_or(mask_blue, mask_red)

# Apply the combined mask to the original blurred image
result = cv2.bitwise_and(img_blur, img_blur, mask=mask_combined)

# Display the original image, blurred image, and result using Matplotlib
plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Blue and Red Colors Extracted'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the original image
img = cv2.imread(r"C:\Users\Keerthana\Desktop\image_new.jpeg")

# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img, (5, 5), 0)

# Convert the blurred image to grayscale
img_gray = cv2.cvtColor(img_blur, cv2.COLOR_BGR2GRAY)

# Apply thresholding
_, binary_thresh = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY)

# Display the original, blurred, and thresholded images using Matplotlib
plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(binary_thresh, cmap='gray')
plt.title('Thresholded Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
